# Base para la simulación

Bueno, llegados a este punto vamos a usar solo 3 columnas: Ciudad, Elección electoral y Votos para la muestra.

```python
v = votesper[votesper['Votos proporcionales'] != 0][['PARROQUIA_CODIGO', 
                                                     'Elección política', 
                                                     'Votos proporcionales']]
v
```

```python
v = v.groupby(by=['PARROQUIA_CODIGO']).size().reset_index(name='Número de personas')
v
```

Para seleccionar muestras de tamañao n, vamos a inspirarnos en el código de la respuesta 5 del siguiente *stackoverflow* https://stackoverflow.com/questions/22472213/python-random-selection-per-group.

Como ejemplo seleccionemos una sola persona por cada ciudad:

```python
# Solo un ajuste antes:

muestra.rename(columns={'Ciudad': 'PARROQUIA_CODIGO'}, inplace=True)

# Seleccionar solo una persona de muestra

n = 1
muestra.groupby(by=['PARROQUIA_CODIGO']).apply(lambda x: x.sample(min(n,len(x)))).reset_index(drop=True)
```

## Algortimo para seleccionar submuestras especificadas de tamaño n y obtener una base para la simulación

**Descripcción**: 

Vamos a filtrar la base *muestra*, obteniendo submuestras para cada ciudad. <br>
Lo que deseamos es que en cada ciudad se elija un número de persona especificadas en la columna *Número de personas a seleccionar* de la tabla *v* aleatoriamente (v['Número de personas a seleccionar']). <br>
Para cada submuestra de ciudad vamos a elegir un número n de persona aleatoriamente.
Vamos a hacer esto 91 veces, ya que hay 91 ciudades. Esta selección significa tomar un número n de personas con  muestra.shape[1] (41) valores de variables.
Finalmente, vamos a juntar estas submustras obtenidas de la selección de el número n de personas en una base llamada df_simulation.


```python
limit = v.shape[0] # Número indicador de filas apra usar en el bucle
Frames = [] # Submuestras a concatenar en una sola base

for i in range(limit):
    df_PARROQUIA_CODIGO = muestra[muestra['PARROQUIA_CODIGO'] == v['PARROQUIA_CODIGO'][i]]
    select = v['Número de personas'][i]
    frame = df_PARROQUIA_CODIGO.groupby(by=['PARROQUIA_CODIGO']).apply(lambda x: x.sample(min(select,len(x)))).reset_index(drop=True)
    Frames.append(frame)

df_simulation = pd.concat(Frames)
df_simulation
```

Será importante verificar que lo que hemos hecho no contenga errores. Para asegurarnos vamos a contar el número de persona que hay en cada submuestra. La siguiente tabla debería ser equivalente a la tabla *v*.

<center><b>Verificación de personas seleccionadas del ENEMDU por parroquia</b></center>

```python
df_simulation.groupby(by=['PARROQUIA_CODIGO'])['Identificador de persona'].size().reset_index(name='Número de personas seleccionadas')
```

Una vez verificado podemos proceder a asignar una elección política aleatoriamente a cada persona de una ciudad.

#### AAG2

```python
r = votesper[votesper['Votos proporcionales'] != 0][['PARROQUIA_CODIGO', 'Elección política', 'Votos proporcionales']]
r.shape
```

```python
lim = r.shape[0]
FRAMES = []

for ner in range(lim):
    row = r.values[ner] # selecciona la fila para repetir el número de veces indicado
    n_repit = r['Votos proporcionales'].values[ner] # selecciona el número de veces indicado
    sigma = np.repeat([r.values[ner]], [n_repit], axis=0)
    k = pd.DataFrame(sigma) # Creo un DataFrame
    FRAMES.append(k) # Agrego el df para concatenarlo luego

df_s = pd.concat(FRAMES)
df_s.rename(columns={0: 'PARROQUIA_CODIGO', 1: 'Elección política', 2: 'Votos proporcionales'}, inplace=True)
df_s
```

### Verifiación
Lo que hemos hecho debería haber repetido la elección política el número de veces el número de votos proporcionales, vamos a crear un conteo de esa repetición que llamaremos *n*.

```python
df_s.groupby(by=['PARROQUIA_CODIGO', 'Elección política']).size().reset_index(name='n').sort_values(['PARROQUIA_CODIGO', 'Elección política'])
```

*n* deberá ser igual a lso votos proporcionales para cada parroquia y elección política.

```python
r.sort_values(['PARROQUIA_CODIGO', 'Elección política'])
```
